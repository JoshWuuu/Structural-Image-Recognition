```python
import numpy as np
import torchvision as tv
import torchvision.transforms as transforms
```

```python
X_train = np.load('Data/Chunks/X_test_1.npy')
```

```python
def recenter(_image):
    image = np.copy(_image)
    image[:, :, 0] -= np.min(image[:, :, 0])
    image[:, :, 1] -= np.min(image[:, :, 1])
    image[:, :, 2] -= np.min(image[:, :, 2])
    return image.astype(np.uint8)
```

```python
toPilTransform = transforms.Compose([
    transforms.Lambda(recenter), 
    transforms.ToPILImage()
])

randomCropTransform = transforms.Compose([
    toPilTransform,
    transforms.RandomHorizontalFlip(1)
])
```
